2. **Претпроцесирање. Конвертирање на атрибутот 'top genre' во нумерички атрибут**. Со цел компатибилност со повеќето алгоритми за класификација кои прифаќаат само нумерички атрибути, во помош на LabelEncoder, правиме асоцијација на жанровите со цели броеви, за да може подоцна да се користи како нумерички атрибут.

```
import numpy as np # linear algebra
import pandas as pd
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/16v9v2020.csv")
df.shape
```

```
df = df.drop("Unnamed: 0", axis="columns")
df.head()
```

```
df['top genre'] = df['top genre'].replace(np.nan,'unknown') 
```

```
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df['genre'] = labelencoder.fit_transform(df['top genre'])
# Viewing few rows of make and its encoded columns
df[['top genre', 'genre']].sample(20)
```

```
df = df.drop(['top genre'], axis = 1)
```
